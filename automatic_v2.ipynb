```python
cpe = pd.read_excel (r'cpe.xlsx')
cpe = pd.DataFrame(cpe)
```

```python
import numpy as np
cpe['inc_Ag_Jul'] = np.where((cpe['vta_202008']- cpe["vta_202007"] > 0) , 1, 0)
cpe['inc_Jul_Jun'] = np.where((cpe['vta_202007']- cpe["vta_202006"] > 0) , 1, 0)
cpe['inc_Jun_May'] = np.where((cpe['vta_202006']- cpe["vta_202005"] > 0) , 1, 0)
cpe['inc_May_Abr'] = np.where((cpe['vta_202005']- cpe["vta_202004"] > 0) , 1, 0)
cpe['inc_Jun_Abr'] = np.where((cpe['vta_202006']- cpe["vta_202004"] > 0) , 1, 0)
cpe['monto_Jun-Abr'] = cpe['vta_202006']- cpe["vta_202004"]
cpe['monto_Ago-Jul'] = cpe['vta_202008']- cpe["vta_202007"]
#cpe['inc_Abr_Mar'] = np.where((cpe['vta_202004']- cpe["vta_202003"] > 0) , 1, 0)
cpe
```

```python
resultado = result3.merge(cpe,on='ruc',how='left')
resultado_df = resultado.merge(dj,on='ruc',how='left')
```

```python
total = resultado_df['monto_Jun-Abr'].sum().round(2)
rucs = resultado_df.query("inc_Jun_Abr == 1")['ruc'].count()
total2 = resultado_df.query("inc_Jun_Abr == 1")['monto_Jun-Abr'].sum().round(2)

print('='*100)
print('El total del Incremento entre Abril y Junio es:' + str(total))
print('El total de RUCs entre con incremento entre Abril y Junio es:'+str(rucs))
print('El total del Incremento entre Abril y Junio y valor 1 en incremento es:'+str(total2))
```

```python
total = resultado_df['monto_Ago-Jul'].sum().round(2)
rucs = resultado_df.query("inc_Ag_Jul == 1")['ruc'].count()
total2 = resultado_df.query("inc_Ag_Jul == 1")['monto_Ago-Jul'].sum().round(2)

print('='*100)
print('El total del Incremento entre Agosto y Julio es:' + str(total))
print('El total de RUCs entre con incremento entre Agosto y Julio es:'+str(rucs))
print('El total del Incremento entre Agosto y Julio y valor 1 en incremento es:'+str(total2))
```
